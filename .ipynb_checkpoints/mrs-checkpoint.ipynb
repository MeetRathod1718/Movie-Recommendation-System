```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].fillna("[]").apply(convert)
```
