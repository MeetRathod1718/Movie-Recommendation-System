```python
import ast

def convert(obj):
    L = []
    try:
        if isinstance(obj, str):  # Ensure obj is a string before parsing
            parsed_obj = ast.literal_eval(obj)  # Convert string to Python object
            for i in parsed_obj:
                if isinstance(i, dict) and 'name' in i:
                    L.append(i['name'])  # Extract keyword names
        elif isinstance(obj, list):  # Already a list
            return obj
    except (ValueError, SyntaxError):
        return []  # Return an empty list if parsing fails

    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
import ast

def convert3(obj):
    L = []
    counter = 0

    if isinstance(obj, str):  # Ensure obj is a string before evaluating
        try:
            obj = ast.literal_eval(obj)  # Convert string representation to a list
        except (ValueError, SyntaxError):
            return []  # Return empty list if parsing fails

    if isinstance(obj, list):  # Ensure obj is a list before iterating
        for i in obj:
            if isinstance(i, dict) and 'name' in i:  # Extract 'name' only if obj is a dict
                if counter < 3:  # Take only first 3 names
                    L.append(i['name'])
                    counter += 1
            elif isinstance(i, str):  # If it's a string (list of names), add directly
                if counter < 3:
                    L.append(i)
                    counter += 1

    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
    return L 
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y = []

    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)   
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
vectors.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
